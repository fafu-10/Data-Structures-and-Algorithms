# Keyboard Shortcuts

1. Select a cell to edit: enter OR double-click
2. Save an edit: shift + enter
3. Create a new cell above: a
4. Create a new cell below: b
5. Convert a code cell to a text cell: m
6. Convert a text cell to a code cell: y
7. Delete a cell: d + d
8. Cut a cell: x
9. Paste a cell: v
10. Undo: z
11. Merge cells: shift to select + M
12. Uncomment/comment code: ctrl + /
13. Multiple line edit: ctrl + click where you want to edit
